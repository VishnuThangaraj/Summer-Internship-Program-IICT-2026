# Library Management System

This notebook implements a Library Management System using Object-Oriented Programming (OOP) principles in Python. It includes classes for managing books, members, and the library transactions (borrowing and returning books).

```python
class Book:
    """
    Represents a book in the library.
    """
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn
        self.is_available = True

    def __str__(self):
        status = "Available" if self.is_available else "Borrowed"
        return f"'{self.title}' by {self.author} (ISBN: {self.isbn}) - [{status}]"
```

```python
class Member:
    """
    Represents a library member who can borrow and return books.
    """
    def __init__(self, name, member_id):
        self.name = name
        self.member_id = member_id
        self.borrowed_books = []

    def borrow_book(self, book):
        if book.is_available:
            book.is_available = False
            self.borrowed_books.append(book)
            print(f"[Success] {self.name} borrowed '{book.title}'.")
            return True
        else:
            print(f"[Failed] '{book.title}' is currently borrowed/unavailable.")
            return False

    def return_book(self, book):
        if book in self.borrowed_books:
            book.is_available = True
            self.borrowed_books.remove(book)
            print(f"[Success] {self.name} returned '{book.title}'.")
            return True
        else:
            print(f"[Failed] {self.name} did not borrow '{book.title}'.")
            return False

    def __str__(self):
        borrowed = ", ".join([f"'{b.title}'" for b in self.borrowed_books]) if self.borrowed_books else "None"
        return f"Member: {self.name} (ID: {self.member_id}) | Borrowed: {borrowed}"
```

```python
class Library:
    """
    Represents the library containing books and registered members.
    """
    def __init__(self, name):
        self.name = name
        self.books = {}      
        self.members = {}    

    def add_book(self, book):
        if book.isbn in self.books:
            print(f"[Notice] Book with ISBN {book.isbn} already exists in '{self.name}'.")
        else:
            self.books[book.isbn] = book
            print(f"[Added] Book '{book.title}' (ISBN: {book.isbn}) to '{self.name}'.")

    def register_member(self, member):
        if member.member_id in self.members:
            print(f"[Notice] Member ID {member.member_id} is already registered.")
        else:
            self.members[member.member_id] = member
            print(f"[Registered] Member '{member.name}' (ID: {member.member_id}).")

    def find_book(self, isbn):
        return self.books.get(isbn, None)

    def find_member(self, member_id):
        return self.members.get(member_id, None)

    def lend_book(self, member_id, isbn):
        member = self.find_member(member_id)
        book = self.find_book(isbn)

        if not member:
            print(f"[Error] Member ID {member_id} not found.")
            return False
        if not book:
            print(f"[Error] Book with ISBN {isbn} not found.")
            return False

        return member.borrow_book(book)

    def return_book(self, member_id, isbn):
        member = self.find_member(member_id)
        book = self.find_book(isbn)

        if not member:
            print(f"[Error] Member ID {member_id} not found.")
            return False
        if not book:
            print(f"[Error] Book with ISBN {isbn} not found.")
            return False

        return member.return_book(book)

    def search_books(self, query):
        query = query.lower()
        results = [book for book in self.books.values() 
                   if query in book.title.lower() or query in book.author.lower() or query in book.isbn.lower()]
        return results

    def list_books(self):
        print(f"\n--- Books in Catalog of '{self.name}' ---")
        if not self.books:
            print("No books available.")
        else:
            for book in self.books.values():
                print(book)

    def list_members(self):
        print(f"\n--- Members of '{self.name}' ---")
        if not self.members:
            print("No members registered.")
        else:
            for member in self.members.values():
                print(member)
```

```python
# Create an instance of the Library
central_library = Library("City Central Library")

# Add some Books
b1 = Book("The Hobbit", "J.R.R. Tolkien", "9780261102217")
b2 = Book("1984", "George Orwell", "9780451524935")
b3 = Book("To Kill a Mockingbird", "Harper Lee", "9780061120084")

print("--- Populating Library ---")
central_library.add_book(b1)
central_library.add_book(b2)
central_library.add_book(b3)

# Register some Members
m1 = Member("Alice Vance", "M101")
m2 = Member("Bob Miller", "M102")

print("\n--- Registering Members ---")
central_library.register_member(m1)
central_library.register_member(m2)

# Initial Display
central_library.list_books()
central_library.list_members()

# Borrowing process
print("\n--- Simulating Borrowing ---")
# Alice borrows 'The Hobbit'
central_library.lend_book("M101", "9780261102217")
# Bob tries to borrow 'The Hobbit' (should fail)
central_library.lend_book("M102", "9780261102217")
# Bob borrows '1984'
central_library.lend_book("M102", "9780451524935")

# Intermediate display
central_library.list_books()
central_library.list_members()

# Returning process
print("\n--- Simulating Returning ---")
# Alice returns 'The Hobbit'
central_library.return_book("M101", "9780261102217")
# Now Bob can borrow 'The Hobbit'
central_library.lend_book("M102", "9780261102217")

# Final Display
central_library.list_books()
central_library.list_members()

# Search functionality
print("\n--- Testing Search ---")
search_results = central_library.search_books("Orwell")
print("Search results for 'Orwell':")
for book in search_results:
    print(book)
```

### Summary of System Logic

The Library Management System is designed using **Object-Oriented Programming (OOP)** concepts, keeping modularity, data encapsulation, and class interaction in mind. It consists of three primary classes:

1. **`Book`**:
   - **Properties**: Stores information about a single book, including its `title`, `author`, `isbn`, and `is_available` (boolean flag to track borrowing status).
   - **Representation**: Implements `__str__` for human-readable output of its title, author, isbn, and status (Available or Borrowed).

2. **`Member`**:
   - **Properties**: Represents a library user with a name, unique ID, and a list of currently borrowed `Book` objects (`borrowed_books`).
   - **Operations**:
     - `borrow_book(book)`: Sets the book's availability to `False` and adds it to the member's list of borrowed books.
     - `return_book(book)`: Sets the book's availability to `True` and removes it from the list of borrowed books.

3. **`Library`**:
   - **Properties**: Serves as the central manager, maintaining catalogs of books (`books` dictionary mapping ISBN to `Book`) and members (`members` dictionary mapping Member ID to `Member`).
   - **Operations**:
     - Book additions (`add_book`) and member registrations (`register_member`).
     - Lookups and transactional tasks (`lend_book` and `return_book`) by mapping strings to internal objects and calling their methods.
     - Search functions to look up books dynamically by title, author, or ISBN.
